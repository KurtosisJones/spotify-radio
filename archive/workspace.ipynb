```python
import numpy as np

def document_vector(model, doc):
    doc = [word for word in doc if word in model.wv.index_to_key]
    if len(doc) > 0:
        return np.mean(model.wv[doc], axis=0)
    else:
        return np.zeros(model.vector_size)
```
